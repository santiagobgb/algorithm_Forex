>><b>1. Estacionariedad:</b>
>> Cuando existe estacionariedad en un serie de tiempo esto indica que las propiedades de la serie no varían con respecto al tiempo. En otras palabras indica que ls características estadísticas de la serie de tiempo serán las mismas en el futuro ycomo en el pasado.

```python
import funciones as fn
df_estacionariedad = fn.df_box_jenkins_estacionariedad()
df_estacionariedad
```

Para la estacionaridad se sacó la media y la varianza e tres diferentes etapas de nuestro indicador. Como se puede apreciar, ni la media ni la varianza son constantes a través del tiempo, por lo que la serie del indicador no es estacionaria.



>><b>2. Estacionalidad:</b>
>> Es prueba mostrará tendencias que dependen de un rango específico de tiempo o periodo.

```python
import funciones as fn
df_estacionalidad = fn.df_box_jenkins_estacionalidad()
df_estacionalidad
```

En estas gráficas se puede apreciar el estimado de la estacionalidad para la serie de tiempo del indicador. 

```python
df_season = fn.df_season()
df_season
```

En esta gráfica se puede apreciar que en el mismo mes pero de cada año existe una ligera similitud mes con més, pero esta no es consecutiva ni entre los años ni entre los meses, por lo que sería difícil decir que esta serie si tiene estacionalidad.

>><b>3. Prueba de Heterocedasticidad (Prueba ARCH):</b>
>> Esta prueba nos indica que la varianza del error es diferente para cada valor de x en una serie de tiempo. Este termino es contrario a la homocedasticidad, la cual implica que, condicionando en las variables explicativas, la varianza del término de error no observado es constante.

```python
import funciones as fn
p_arch = fn.df_parch()
p_arch
```

Dado que el P-value es menor a 0.05 rechazamos la hipótesis nula, nuestra serie es heterocedástica. 

>><b>4. Autocorrelación & Autocorrelación Parcial:</b>
>> El componente de autocorrelación busca patrones repetitivos dentro de una serie de tiempo (sin rezagos). Ahora, cuando se trata de autocorrelación parcial, ésta se encarga de definir la correlación parcial de la serie de tiempo estacionaria con sus propios valores rezagados.

```python
fn.df_box_jenkins_acf()
fn.df_box_jenkins_pacf()
```

La grafica de autocorrelación simple nos indica que existe una relación negativas de aproximadamente 37% entre el valor en tiempo actual y el valor de un periodo anterior. La autocorrelación parcial dos periodos antes del tiempo real es de aproximadamente 21%. Esto nos indica que son muy pocoslos rezagos con los que se pueden trabajar para las posibles predicciones en en el comportamiento de esta serie, además de que la relación entre los datos sucesivos es muy baja. 

>><b>5. Detección de Atípicos:</b>
>> La detección de atípicos implica determinar la ubicación, tipo y magnitud de todos los valores atípicos presentes. Se utlizará la gráfica Boxplot para identificar valores atípicos en la serie.

```python
import funciones as fn
%matplotlib inline
df_atip = fn.df_boxplot()
df_atip
#no se imprimen aquí las gráficas wtf
```

Para los valores atípicos se utilizó un Boxplot (solo para el Actual), este lo que hace es utilizar las fórmulas de Q1 -1.5*IQR y Q3 + 1.5*IQR donde IQR es el rango de cada cuartil y Qi (donde i es 1 o 3). Estos datos son atípicos ya sus valores no se ven reflejados por las características de la serie de tiempo.
